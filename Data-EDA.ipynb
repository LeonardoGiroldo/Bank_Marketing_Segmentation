```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set()
import matplotlib.pyplot as plt
```

```python
creditcard_df_cleaned = pd.read_csv('Data-Cleaned.csv')
creditcard_df_cleaned
```

```python
creditcard_df_cleaned.drop(['Unnamed: 0'], axis = 'columns', inplace = True)
```

```python
creditcard_df_cleaned.head()
```

```python
n = len(creditcard_df_cleaned.columns)
n
```

```python
plt.figure(figsize=(10,50))
for i in range(n):
  plt.subplot(17, 1, i+1)
  sns.distplot(creditcard_df_cleaned[creditcard_df_cleaned.columns[i]])
  plt.title(creditcard_df_cleaned.columns[i])

plt.tight_layout()
```

```python
correlations = creditcard_df_cleaned.corr()
```

```python
plt.figure(figsize = (20, 20))
sns.heatmap(correlations, annot = True)
```
